# Kelly Criterion in a Binomial Setting

```python
import math
import time
import numpy as np
import pandas as pd
import datetime as dt
import cufflinks as cf
from pylab import mpl, plt
```

```python
np.random.seed(1000)
plt.style.use('seaborn')
mpl.rcParams['font.family'] = 'serif'
%matplotlib inline
```

```python
p = 0.55
```

```python
f = p -(1 - p)
```

```python
f
```

```python
I = 50
```

```python
n = 100
```

```python
def run_simulation(f):
  c = np.zeros((n, I))
  c[0] = 100
  for i in range(I):
    for t in range(1, n):
      o = np.random.binomial(1, p)
      if o > 0:
        c[t, i] = (1 + f) * c[t - 1, i]
      else:
        c[t, i] = (1 - f) * c[t - 1, i]
  return c
```

```python
c_1 = run_simulation(f)
```

```python
c_1.round(2)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(c_1, 'b', lw=0.5)
plt.plot(c_1.mean(axis=1), 'r', lw=2.5)
```

```python
c_2 = run_simulation(0.05)
```

```python
c_3 = run_simulation(0.25)
```

```python
c_4 = run_simulation(0.5)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(c_1.mean(axis=1), 'black', label='$f^*=0.1$')
plt.plot(c_2.mean(axis=1), 'pink', label='$f=0.05$')
plt.plot(c_3.mean(axis=1), 'green', label='$f=0.25$')
plt.plot(c_4.mean(axis=1), 'white', label='$f=0.5$')
plt.legend(loc=0)
```

```python
raw = pd.read_csv('tr_eikon_eod_data.csv', index_col=0, parse_dates=True)
```

```python
symbol = '.SPX'
```

```python
data = pd.DataFrame(raw[symbol])
```

```python
data['returns'] = np.log(data / data.shift(1))
```

```python
data.dropna(inplace=True)
```

```python
data.head()
```

```python
data.tail()
```

```python
data.describe().round(2)
```

```python
data.info()
```

```python
mu = data.returns.mean() * 252
mu
```

```python
sigma = data.returns.std() * 252 ** 0.5
sigma
```

```python
r = 0.0
```

```python
f = (mu - r) / sigma ** 2
f
```

```python
equs = []
```

```python
def kelly_strategy(f):
  global equs
  equ = 'equity_{:.2f}'.format(f)
  equs.append(equ)
  cap = 'capital_{:.2f}'.format(f)
  data[equ] = 1
  data[cap] = data[equ] * f
  for i, t in enumerate(data.index[1:]):
    t_1 = data.index[i]
    data.loc[t, cap] = data[cap].loc[t_1] * \
                        math.exp(data['returns'].loc[t])
    data.loc[t, equ] = data[cap].loc[t] - \
                        data[cap].loc[t_1] + \
                          data[equ].loc[t_1]
    data.loc[t, cap] = data[equ].loc[t] * f
```

```python
kelly_strategy(f * 0.5)
```

```python
kelly_strategy(f * 0.66)
```

```python
kelly_strategy(f)
```
